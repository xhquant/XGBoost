```python
from sklearn.model_selection import KFold, train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn import datasets
from sklearn.metrics import roc_auc_score

import xgboost as xgb
```

```python
print("二分类，乳腺癌数据")
cancer = datasets.load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
kf = KFold(n_splits=5, shuffle=True)
i = 0
```

```python
for train_idx, test_idx in kf.split(X):
    model = xgb.XGBClassifier(max_depth=6, 
                              max_leaves=50, 
                              max_bin=256,
                              n_estimators=80, 
                              learning_rate=0.05,
                              verbosity=1,
                              objective="binary:logistic",
                              booster="gbtree",
                              subsample=0.8,
                              eval_metric='auc',
                              early_stopping_rounds=10).fit(X=X[train_idx], y=y[train_idx], eval_set=[(X[train_idx], y[train_idx]), (X[test_idx], y[test_idx])])
    y_pred = model.predict(X[test_idx])
    labels = y[test_idx]
    print("kfold-%d AUC为: %f" % (i, roc_auc_score(y_true=labels,y_score=y_pred)))
    i += 1
    
booster = model.get_booster()
booster.save_model("model/scikit_learn.xgb")
```
